     [12, 30, 30, 9]

İpucu : map() fonksiyonunu kullanmaya çalışın.

````python
# 1.Problemin Çözümünü Buraya Yazınız

def dortgen():
    list = [(3,4),(10,3),(5,6),(1,9)]
    new_list = []
    s = 0
    while s < 4:
        
        alan = list[s][0] * list[s][1]
        s += 1
        new_list.append(alan)
    print(new_list)
dortgen() 
````
