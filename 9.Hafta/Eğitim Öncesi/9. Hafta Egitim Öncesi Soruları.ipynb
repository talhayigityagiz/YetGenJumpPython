# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız

class vehicle:
    def __init__(self,line,capacity,type):
        self.line = line
        self.capacity = capacity
        self.type = type 
        print("Araç Bilgisi Hazır.")
    def intro(self):
        print(self.line, self.capacity, self.type)
        
class bus(vehicle):
    def __init__(self,line,capacity,type):
        super(). __init__(line,capacity,type)
    print("Otobüs Bilgisi hazır.")
    
    def run(self):
        print(f"Otobüs yolcu limiti : {self.capacity}")
        
b1 = bus("B250","50","Bus")
b1.run()
````
