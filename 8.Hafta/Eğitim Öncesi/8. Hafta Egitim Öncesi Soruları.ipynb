# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
urun_ismi = input("Ürünün ismini giriniz: ")
urun_fiyati = int(input("Ürünün fiyatını giriniz: "))

def kayit():
    with open("urunler.txt", "r+", encoding="utf-8") as file:
        file.write(f"Ürünün İsmi = {urun_ismi} \n")
        file.write(f"Ürünün Fiyatı = {urun_fiyati} \n")
        for i in file:
            print(i)
    
kayit()
````
